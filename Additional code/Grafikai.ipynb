```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
classes_grouped = [['YOLOv4-tiny 1 versija', 'YOLOv4-tiny 2 versija', 'YOLOv4 (paskutinės apmokymo dalies)'], [87.58, 78.54, 96.01], [85.88, 87.40, 95.10], [35.76, 72.41, 95.09]]
classes_names = ['Metodai', 'Teisingai dėvima veido kaukė', 'Nedėvima veido kaukė', 'Neteisingai dėvima veido kaukė']
```

```python
df = pd.DataFrame(classes_grouped, index=classes_names).T
df
```

```python
values = [87.58, 85.88, 35.76] + [78.54, 87.40, 72.41] + [96.01, 95.10, 95.09]
labels = ['87.58 %', '85.88 %', '35.76 %', '78.54 %', '87.40 %', '72.41 %', '96.01 %', '95.10 %', '95.09 %']
rows = [-0.23, -0.06, 0.10, 0.77, 0.935, 1.10, 1.775, 1.94, 2.1]

df.plot(x='Metodai',
        y=['Teisingai dėvima veido kaukė', 'Nedėvima veido kaukė', 'Neteisingai dėvima veido kaukė'],
        kind='bar',
        figsize=(20,8),
        colormap='copper',
        rot=0)

plt.ylabel('Vidutinis tikslumas', fontsize=20)
plt.xlabel('Metodai', fontsize=20)

for i in range(len(values)):
    plt.text(x = rows[i] , y = values[i]-4, s = labels[i], size = 12, c='white')
```
